```python
from scipy.io.wavfile import read
import matplotlib.pyplot as plt
input_data = read("/content/beat.wav")
audio = input_data[1]
plt.plot(audio[:])
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.show()
```

```python
import librosa as lib
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft,fftfreq

s, sr = lib.load('/content/piano-C5.wav')
x=fft(s)
size=len(x)
y=fftfreq(size,1/sr)
plt.plot(np.abs(y), np.abs(x))
plt.show()

```
