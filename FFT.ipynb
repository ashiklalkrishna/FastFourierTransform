```python
import librosa as lib
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft,fftfreq

s, sr = lib.load('/content/trumpet-C5.wav')
f=fft(s)
size=len(f)
freq=fftfreq(size,1/sr)

max_index=np.argmax(np.abs(f))                 #To find peak frequency
peak_freq=(np.abs(freq))[max_index]
m=np.max(np.abs(f))

fig = plt.figure(figsize=(10, 5))
plt.plot(np.abs(freq),np.abs(f), color='r')
plt.text (peak_freq, m, f" {int(peak_freq)}Hz", color='black')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Fourier transform of a trumpet C5 sample')
plt.show()
```
