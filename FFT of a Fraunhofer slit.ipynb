```python
import numpy as np
import matplotlib.pyplot as plt

slit = np.zeros((200, 200))
slit[90:110, 90:110] = 1

fig, (img_slit, img_fft) = plt.subplots(1, 2, figsize=(10,5))
img_slit.imshow(slit, cmap='gray')

fft_slit = np.fft.fft2(slit)
fft_shift = np.fft.fftshift(fft_slit)

img_fft.imshow(np.abs(fft_shift), cmap='gray')

plt.show()
```
