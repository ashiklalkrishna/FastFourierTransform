```python
#2D FFT

import numpy as np
import matplotlib.pyplot as plt

img = plt.imread('/content/crop_cos_ver_25_hann.png')
img_gray = np.mean(img, axis=2)   #Convert to greyscale

plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.show()

fft_img = np.fft.fft2(img_gray)
fft_img_shifted = np.fft.fftshift(fft_img)   #Shift zero-frequency component to the center

img_mag= np.abs(fft_img_shifted)    #magnitude of fft

plt.imshow(np.log10(img_mag), cmap='gray')
plt.axis('off')
plt.show()
```

```python
#APPLY HANNING WINDOW

import numpy as np
from scipy.signal import hann
import matplotlib.pyplot as plt
from PIL import Image

image = Image.open('/content/crop_fran_sqr_nohann.png')
image_array = np.array(image)

plt.axis('off')

window_2d = np.outer(hann(image_array.shape[0]), hann(image_array.shape[1]))
image_array = image_array * window_2d[:, :, np.newaxis]

img_gray = np.mean(image_array, axis=2)

plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
#FFT of a FFT

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/fft_fran_circ.png', cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)

fft = np.fft.fft2(img)

fft_shift = np.fft.fftshift(fft)

img_mag = 20*np.log(np.abs(fft_shift)+ 1e-9)
cv2_imshow(img_mag)
```
