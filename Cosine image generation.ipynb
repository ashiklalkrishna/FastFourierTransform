```python
import numpy as np
import matplotlib.pyplot as plt

img_size = 256
f = 100

x, y = np.meshgrid(np.linspace(-1, 1, img_size), np.linspace(-1, 1, img_size))
cos_hor = np.cos(2 * np.pi * f * x)

plt.imshow(cos_hor, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

img_size = 256
f = 20
angle = (np.pi/3)

x, y = np.meshgrid(np.linspace(-1, 1, img_size), np.linspace(-1, 1, img_size))
cos_60 = np.cos(2 * np.pi * f * (x * np.cos(angle) + y * np.sin(angle)))

plt.imshow(cos_60, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

img_size = 256
f = 10
angle = (np.pi/3)

x, y = np.meshgrid(np.linspace(-1, 1, img_size), np.linspace(-1, 1, img_size))
cos_v = np.cos(2 * np.pi * f * (x * np.cos(angle) + y * np.sin(angle)))
cos_h = np.cos(2 * np.pi * f * (x * np.cos(angle) - y * np.sin(angle)))
cos_img = cos_v*cos_h

plt.imshow(cos_img, cmap='gray')
plt.axis('off')
plt.show()
```
